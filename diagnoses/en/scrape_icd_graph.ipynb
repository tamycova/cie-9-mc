```python
import requests
from time import sleep
```

```python
level = requests.get('https://icd.who.int/browse10/2008/en/JsonGetRootConcepts?useHtml=false').json()
level = [("head", i) for i in level]
nodes = dict()
while level:
    new_level = []
    for parent, node in level:
        _id = node['ID']
        print(_id)
        _label = node['label']
        _leaf = node['isLeaf']
        _depth = node['averageDepth']
        if _id in nodes:
            print("SEEN", _id)
            raise Exception
        nodes[_id] = {"label": _label, "leaf" : _leaf, "depth" : _depth, "parent": parent}
        if not _leaf:
            children = requests.get('https://icd.who.int/browse10/2008/en/JsonGetChildrenConcepts?ConceptId={}&useHtml=false&showAdoptedChildren=true'.format(_id)).json()
            for child in children:
                new_level.append((_id, child))
    level = new_level        
```

```python
from json import dump 

with open("scraped_graph.json", "w") as f:
    dump(nodes, f)
```
